### 1. Import libraries 

```python
import requests
from requests.exceptions import HTTPError

import re
```

### 2. Download Webpage

`dagobah.net` uses index mode to list the pages, and all the swf files are fixed. We can use a systematic way to capture all the page information.

```python
async def download(start=1, max_page=276):   # Maximum Page is 276
    for i in range(start,max_page+1):
        for url in ['https://dagobah.net?page=' + str(i)]:
            try:
                response = requests.get(url)
                response.raise_for_status()
            except HTTPError as http_err:
                print(f'HTTP error occurred: {http_err}')  # Python 3.6
            except Exception as err:
                print(f'Other error occurred: {err}')  # Python 3.6
            else:
                print('Page ' + str(i) + ' Success!')
                await keyword_filter(response.text)
```

### 3. Keyword filter

```python
# keyword filter (bad design)
keyword1 = "onmouseout="
keyword2 = ").href="
keyword3 = "document.getElementById("
```

```python
def slicer(my_str,sub):
    index=my_str.find(sub)
    if index !=-1 :
        return my_str[index:] 
    else:
        raise Exception('Sub string not found!')
```

```python
async def keyword_filter(response_text):
    print("Keyword filter processing...")
    for line in response_text.splitlines():
        sentence = line.strip()
        if all(k in sentence for k in [keyword1, keyword2, keyword3]):
            try:
                combined = slicer(sentence, '/flash/') + '\n'
                replaced = combined.replace("/flash/", "")
                filtered = replaced.replace("'", "")
                await download_resources(filtered.rstrip('\n'))
                print(filtered)
            except Exception:
                continue
```

### 4. Download swf

```python
async def download_resources(path):
    print("Downloading resources " + path + "...")
    file2 = open('result_swf' + path , "wb")
    for url in ['https://dagobah.net/flashswf/' + path]:
        try:
            response = requests.get(url)
            print("Status code: ", response.status_code)
            response.raise_for_status()
        except HTTPError as http_err:
            print(f'HTTP error occured: {http_err}')
        except Exception as err:
            print(f'Other error occured: {err}')
        finally:
            print(path + ' downloaded... success!')
            file2.write(response.content)
```
